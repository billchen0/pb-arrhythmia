```python
from pathlib import Path
import matplotlib.pyplot as plt
from lxml import etree
```

```python
fda_path = Path("/media/nvme1/pbecg-data/fda/2018/R^ECG^F_0^0_5692915810^008^STUDY START_20180216103233_20180804101143.xml")
```

```python
tree = etree.parse(fda_path)
root = tree.getroot()
```

```python
# Extract ECG waveform data from digits tag
namespace = '{urn:hl7-org:v3}'
digits_elements = root.findall(f".//{namespace}digits")
```

```python
ecg_data_txt = digits_elements[0].text
ecg_data = [int(x) for x in ecg_data_txt.split()]

plt.figure(figsize=(12, 6))
plt.plot(ecg_data[:6000])
```
